```python
# TOP 10 STORES WITH HIGH SALES 
```

```python
top_stores_by_company = pd.DataFrame(columns=['STORECODE', 'CMP', 'VALUE'])


unique_storecodes = data['STORECODE'].unique()

fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(16, 8))
fig.suptitle('Top 10 Stores with Highest Sales Amounts for Each Company')


for i, storecode in enumerate(unique_storecodes[:10]): 
    store_data = data[data['STORECODE'] == storecode]
    top_companies = store_data.groupby('CMP')['VALUE'].sum().reset_index()

    top_companies = top_companies.sort_values(by='VALUE', ascending=False).head(10)

    top_companies['STORECODE'] = storecode

    
    top_stores_by_company = pd.concat([top_stores_by_company, top_companies], ignore_index=True)

    
    row, col = i // 5, i % 5
    ax = axs[row, col]
    sns.barplot(x='VALUE', y='CMP', data=top_companies, ax=ax, color='green')
    ax.set_title(f'Store {storecode}')
    ax.set_xlabel('VALUE')
    ax.set_ylabel('Company')


plt.tight_layout(rect=[0, 0.10, 1, 0.95])
plt.show()
```

```python
#DAILY SALES ANALYSIS FOR EACH STORE
```

```python
for i, storecode in enumerate(unique_storecodes[:10]): 
    store_data = data[data['STORECODE'] == storecode]
    daily_sales = store_data.groupby("QTY")["VALUE"].sum().reset_index()

    plt.figure(figsize=(20, 5))
    sns.barplot(x=daily_sales['QTY'], y=daily_sales['VALUE']).axhline(daily_sales['VALUE'].mean(), color='red')
    plt.title("Daily Sales Analysis - Store: " + str(storecode))
    plt.xlabel("Day")
    plt.ylabel("Total Sales")
    plt.grid(axis='y', color='black')
    plt.show()
```

```python
# TOP 25 CATEGORIES THAT SOLD
```

```python
top_items_sales = data.groupby('GRP').sum().sort_values("VALUE",ascending=False)[['QTY','MONTH','VALUE']][0:25]
plt.figure(figsize=(12,8))
sns.barplot(x='VALUE',y=top_items_sales.index,data=top_items_sales)
plt.title("Sales from Top 25 Categories")
plt.xlabel("Sales")
plt.ylabel("Categories")
plt.grid(axis='x',color='black');
```

```python
# TOP 25 CATEGORIES SOLD IN UNITS
```

```python
top_items = data.groupby('GRP').sum().sort_values("QTY",ascending=False)[['QTY','VALUE']][0:25]
plt.figure(figsize=(12,8))
sns.barplot(x='QTY',y=top_items.index,data=top_items)
plt.title("Number of Units Sold by Top 25 Categories")
plt.xlabel("Number of Units")
plt.ylabel("Categories")
plt.grid(axis='x',color='red')
```
